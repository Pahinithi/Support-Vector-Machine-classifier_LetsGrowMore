```python
from sklearn.datasets import load_iris

iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)
```

```python
print("Unique classes in y_train:", np.unique(y_train))
```

```python
# import SVC classifier
from sklearn.svm import SVC

# import metrics to compute accuracy
from sklearn.metrics import accuracy_score

# instantiate classifier with default hyperparameters
svc=SVC() 

# fit classifier to training set
svc.fit(X_train,y_train)

# make predictions on test set
y_pred=svc.predict(X_test)

# compute and print accuracy score
print('Model accuracy score with default hyperparameters: {0:0.2f}'. format(accuracy_score(y_test, y_pred)))
```

# Balancing Act: Navigating Overfitting and Underfitting in Machine Learning Models

Overfitting happens when a model does really well on the training data but not so well on new, unseen data. This occurs when the model is too complicated and picks up on random or specific patterns in the training data that don't apply well to new data. On the other hand, underfitting occurs when a model doesn't perform well on either the training or test data, suggesting it's too simple to understand the underlying patterns.

Ideally, you want the accuracy of the test set to be close to the accuracy of the training set. If the model generalizes well, both accuracies should be similar. If the test-set accuracy significantly drops compared to the train-set accuracy, it may mean overfitting. The scores here are similar and satisfactory.
